# Divergence

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vector field F with y set to a constant value
def F(x, z):
    return np.array([x**2 + 1, -4 * x * 1 * z])  # y is set to 1

# Define the divergence function
def divergence(x, z):
    return 2 * x * (z - 1)**2

# Create a grid of points in the x-z plane
x = np.linspace(-2, 2, 20)
z = np.linspace(-2, 2, 20)
X, Z = np.meshgrid(x, z)

# Compute the divergence and the vector field
D = divergence(X, Z)
U, V = F(X, Z)

# Normalize the arrows for better visualization
magnitude = np.sqrt(U**2 + V**2)
U_normalized = U / magnitude
V_normalized = V / magnitude

# Adjust arrows based on divergence sign
arrow_scaling = np.where(D > 0, 1, -1)  # Scale arrows based on divergence sign

# Plotting
plt.figure(figsize=(12, 8))

# Create a quiver plot for the vector field
quiver = plt.quiver(X, Z, U_normalized * arrow_scaling, V_normalized * arrow_scaling, 
                     color='blue', scale=50, pivot='middle', alpha=0.8)


# Add labels and title
plt.title('Vector Field with Divergence Behavior on the X-Z Plane')
plt.xlabel('X-axis')
plt.ylabel('Z-axis')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Add color bar for divergence
plt.colorbar(contour, label='Divergence of F = 2x(z - 1)^2')

plt.grid()
plt.show()
```

# Gradient

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f
def f(x, y):
    return x * y + x + y

# Define the gradient of f
def gradient(x, y):
    df_dx = y + 1  # f/x
    df_dy = x + 1  # f/y
    return np.array([df_dx, df_dy])

# Create a grid of points in the xy-plane
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# Compute the gradient at each point
U, V = gradient(X, Y)

# Plotting
plt.figure(figsize=(10, 8))

# Create a quiver plot for the gradient
plt.quiver(X, Y, U, V, color='blue', scale=60)

# Add labels and title
plt.title('Gradient Field of f(x, y) = xy + x + y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')

plt.show()
```

# Curl

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the curl of F in the xy-plane at z = 0
def curl_F(x, y):
    curl_x = -x * y  # -xy
    curl_y = x * y   # xy
    return curl_x, curl_y

# Create a grid of points in the xy-plane
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Compute the curl at each point
curl_x, curl_y = curl_F(X, Y)

# Plotting
plt.figure(figsize=(10, 8))

# Create a quiver plot for the curl vector field
plt.quiver(X, Y, curl_x, curl_y, color='blue', scale=40)

# Add labels and title
plt.title('Curl of the Vector Field F = <xyz, xyz, xyz> in the XY-Plane (z = 0)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')

plt.show()
```
